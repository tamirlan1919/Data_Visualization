```python
fig = plt.figure(figsize=(25, 10))
ax = fig.add_subplot(111)
plt.plot(df['Entity'],df['Life expectancy'],'r--')
plt.grid()
ax.set_title('График по продолжительности жизни')
ax.set_xlabel('Страны')
ax.set_ylabel('Возраст')
plt.show()
```

```python
fig = plt.figure(figsize=(25, 10))
plt.barh(df['Entity'],df['Life expectancy'])
plt.show()
```

```python
df.hist(figsize = (20,15), edgecolor = 'black',bins = 10,color = 'green')
```

```python
countries=df.Entity.unique()

for i in countries:
    ax = plt.axes()
    ax.set_facecolor("#212225")
# OR
    ax.set(facecolor = "#212225")
    data=df[df.Entity==i]
    plt.xlabel('YEAR`S')
    plt.ylabel('Life expectancy')
    plt.title(i)
    plt.grid()
    plt.plot(data.Year,data['Life expectancy'],c='b')
    plt.show()
```

```python
life_expectancy = df.groupby('Entity').mean().reset_index()
life_expectancy
```

```python
import plotly.express as px
fig = px.scatter(df, x = "Year", y= "Life expectancy", color = "Entity",title = '<b> Рост продолжительности жизни </b>')
fig.show()
```

```python
px.scatter(df, x="Year", y="Life expectancy", animation_frame="Year", animation_group="Страны",size="Life expectancy", color="Entity", hover_name="Entity",
            log_x=True,size_max=55, range_x=[1800,2017], range_y=[1,110])
```

```python
px.bar(df, x="Entity", y="Life expectancy", color="Entity",
animation_frame="Year", animation_group="Entity", range_y=[0,60])
```

```python
df = px.data.gapminder()
fig = px.scatter_geo(df, locations="iso_alpha",
                     color="continent", # which column to use to set the color of markers
                     hover_name="country", # column added to hover information
                     size="pop", # size of markers
                     projection="orthographic")
fig
```
